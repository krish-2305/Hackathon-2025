```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import math

class CircularDefectDetector:
    def __init__(self):
        self.min_radius = 50
        self.max_radius = 500
        self.defect_threshold = 0.02  # 2% deviation from expected radius
        
    def preprocess_image(self, image):
        """Preprocess the image for better circle detection"""
        if len(image.shape) == 3:
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        else:
            gray = image.copy()
        
        # Apply Gaussian blur to reduce noise
        blurred = cv2.GaussianBlur(gray, (5, 5), 1)
        
        # Apply adaptive thresholding for better edge detection
        binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                     cv2.THRESH_BINARY, 11, 2)
        
        return gray, binary
    
    def detect_circles(self, image):
        """Detect inner and outer circles using HoughCircles"""
        gray, binary = self.preprocess_image(image)
        
        # Edge detection
        edges = cv2.Canny(binary, 50, 150, apertureSize=3)
        
        # Detect circles using HoughCircles
        circles = cv2.HoughCircles(
            edges,
            cv2.HOUGH_GRADIENT,
            dp=1,
            minDist=100,
            param1=50,
            param2=30,
            minRadius=self.min_radius,
            maxRadius=self.max_radius
        )
        
        if circles is not None:
            circles = np.round(circles[0, :]).astype("int")
            # Sort circles by radius to get inner and outer
            circles = sorted(circles, key=lambda x: x[2])
            
            if len(circles) >= 2:
                inner_circle = circles[0]  # Smaller radius
                outer_circle = circles[-1]  # Larger radius
                return inner_circle, outer_circle
            elif len(circles) == 1:
                # If only one circle detected, estimate the other
                detected = circles[0]
                center_x, center_y, radius = detected
                
                # Estimate inner/outer based on the binary image
                inner_radius = int(radius * 0.6)  # Estimate
                outer_radius = int(radius * 1.2)  # Estimate
                
                return (center_x, center_y, inner_radius), (center_x, center_y, outer_radius)
        
        return None, None
    
    def extract_ring_boundary(self, image, inner_circle, outer_circle):
        """Extract the boundary points of the ring for defect analysis"""
        gray, _ = self.preprocess_image(image)
        
        # Use the average center from both circles
        center_x = int((inner_circle[0] + outer_circle[0]) / 2)
        center_y = int((inner_circle[1] + outer_circle[1]) / 2)
        inner_radius = inner_circle[2]
        outer_radius = outer_circle[2]
        
        # Extract boundary points using polar coordinates
        angles = np.linspace(0, 2*np.pi, 360)
        inner_boundary_points = []
        outer_boundary_points = []
        
        for angle in angles:
            # Sample points along radial lines
            for r in range(max(1, inner_radius-10), inner_radius+10):
                x = int(center_x + r * np.cos(angle))
                y = int(center_y + r * np.sin(angle))
                
                if 0 <= x < gray.shape[1] and 0 <= y < gray.shape[0]:
                    # Find the actual boundary by looking for intensity changes
                    if self.is_boundary_point(gray, x, y):
                        inner_boundary_points.append((x, y, angle, r))
                        break
            
            for r in range(outer_radius-10, min(gray.shape[0]//2, outer_radius+10)):
                x = int(center_x + r * np.cos(angle))
                y = int(center_y + r * np.sin(angle))
                
                if 0 <= x < gray.shape[1] and 0 <= y < gray.shape[0]:
                    if self.is_boundary_point(gray, x, y):
                        outer_boundary_points.append((x, y, angle, r))
                        break
        
        return inner_boundary_points, outer_boundary_points, (center_x, center_y)
    
    def is_boundary_point(self, gray_image, x, y, threshold=30):
        """Check if a point is on a boundary based on gradient"""
        if x <= 1 or x >= gray_image.shape[1]-2 or y <= 1 or y >= gray_image.shape[0]-2:
            return False
        
        # Calculate gradient magnitude
        gx = int(gray_image[y, x+1]) - int(gray_image[y, x-1])
        gy = int(gray_image[y+1, x]) - int(gray_image[y-1, x])
        gradient_mag = np.sqrt(gx*gx + gy*gy)
        
        return gradient_mag > threshold
    
    def detect_defects(self, boundary_points, expected_radius, center):
        """Detect defects by analyzing deviation from expected circular boundary"""
        if not boundary_points:
            return []
        
        defects = []
        center_x, center_y = center
        
        # Calculate expected vs actual radius for each point
        for x, y, angle, actual_radius in boundary_points:
            # Calculate distance from center
            distance_from_center = np.sqrt((x - center_x)**2 + (y - center_y)**2)
            
            # Calculate deviation from expected radius
            deviation = abs(distance_from_center - expected_radius) / expected_radius
            
            if deviation > self.defect_threshold:
                defect_type = self.classify_defect(deviation, distance_from_center, expected_radius)
                defects.append({
                    'position': (x, y),
                    'angle': angle,
                    'deviation': deviation,
                    'type': defect_type,
                    'severity': min(deviation * 10, 1.0)  # Normalize severity
                })
        
        # Cluster nearby defects
        clustered_defects = self.cluster_defects(defects)
        return clustered_defects
    
    def classify_defect(self, deviation, actual_radius, expected_radius):
        """Classify defect as 'cut' or 'flash' based on characteristics"""
        if actual_radius < expected_radius:
            return 'cut'  # Material missing (inward deviation)
        else:
            return 'flash'  # Extra material (outward deviation)
    
    def cluster_defects(self, defects):
        """Group nearby defects using DBSCAN clustering"""
        if len(defects) < 2:
            return defects
        
        # Extract positions for clustering
        positions = np.array([defect['position'] for defect in defects])
        
        # Apply DBSCAN clustering
        clustering = DBSCAN(eps=20, min_samples=2).fit(positions)
        labels = clustering.labels_
        
        clustered_defects = []
        unique_labels = set(labels)
        
        for label in unique_labels:
            if label == -1:  # Noise points
                mask = labels == label
                for i, defect in enumerate(defects):
                    if mask[i]:
                        clustered_defects.append(defect)
            else:
                # Group defects in the same cluster
                cluster_defects = [defects[i] for i, l in enumerate(labels) if l == label]
                
                # Create a representative defect for the cluster
                avg_x = np.mean([d['position'][0] for d in cluster_defects])
                avg_y = np.mean([d['position'][1] for d in cluster_defects])
                max_deviation = max([d['deviation'] for d in cluster_defects])
                common_type = max(set([d['type'] for d in cluster_defects]), 
                                key=[d['type'] for d in cluster_defects].count)
                
                clustered_defects.append({
                    'position': (int(avg_x), int(avg_y)),
                    'deviation': max_deviation,
                    'type': common_type,
                    'severity': min(max_deviation * 10, 1.0),
                    'cluster_size': len(cluster_defects)
                })
        
        return clustered_defects
    
    def process_image(self, image_path):
        """Main processing function"""
        # Load image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Could not load image: {image_path}")
            return None
        
        print(f"Processing: {image_path}")
        
        # Step 1: Detect circles
        inner_circle, outer_circle = self.detect_circles(image)
        
        if inner_circle is None or outer_circle is None:
            print("Could not detect circles in the image")
            return None
        
        print(f"Inner circle: center=({inner_circle[0]}, {inner_circle[1]}), radius={inner_circle[2]}")
        print(f"Outer circle: center=({outer_circle[0]}, {outer_circle[1]}), radius={outer_circle[2]}")
        
        # Step 2: Extract ring boundaries
        inner_boundary, outer_boundary, center = self.extract_ring_boundary(
            image, inner_circle, outer_circle)
        
        # Step 3: Detect defects on both boundaries
        inner_defects = self.detect_defects(inner_boundary, inner_circle[2], center)
        outer_defects = self.detect_defects(outer_boundary, outer_circle[2], center)
        
        all_defects = inner_defects + outer_defects
        
        # Step 4: Visualize results
        result_image = self.visualize_results(image, inner_circle, outer_circle, 
                                            all_defects, center)
        
        # Print results
        if all_defects:
            print(f"Found {len(all_defects)} defect(s):")
            for i, defect in enumerate(all_defects):
                print(f"  Defect {i+1}: {defect['type']} at {defect['position']}, "
                      f"severity: {defect['severity']:.2f}")
        else:
            print("No defects detected - Part appears to be good")
        
        return {
            'image': image,
            'result_image': result_image,
            'inner_circle': inner_circle,
            'outer_circle': outer_circle,
            'defects': all_defects,
            'status': 'defective' if all_defects else 'good'
        }
    
    def visualize_results(self, image, inner_circle, outer_circle, defects, center):
        """Create visualization of detection results"""
        result = image.copy()
        
        # Draw detected circles
        cv2.circle(result, (inner_circle[0], inner_circle[1]), inner_circle[2], (0, 255, 0), 2)
        cv2.circle(result, (outer_circle[0], outer_circle[1]), outer_circle[2], (0, 255, 0), 2)
        
        # Draw center
        cv2.circle(result, center, 5, (255, 0, 0), -1)
        
        # Draw defects
        for i, defect in enumerate(defects):
            x, y = defect['position']
            color = (0, 0, 255) if defect['type'] == 'cut' else (255, 0, 255)  # Red for cuts, Magenta for flashes
            
            # Draw defect location
            cv2.circle(result, (x, y), 8, color, 2)
            
            # Add label
            label = f"{defect['type']}"
            cv2.putText(result, label, (x+10, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
        
        return result

# Example usage and testing
def main():
    detector = CircularDefectDetector()
    
    # Process each image
    image_paths = []  # Update with actual paths
    
    for image_path in image_paths:
        try:
            result = detector.process_image(image_path)
            if result:
                # Display results
                plt.figure(figsize=(15, 5))
                
                plt.subplot(1, 3, 1)
                plt.imshow(cv2.cvtColor(result['image'], cv2.COLOR_BGR2RGB))
                plt.title('Original Image')
                plt.axis('off')
                
                plt.subplot(1, 3, 2)
                plt.imshow(cv2.cvtColor(result['result_image'], cv2.COLOR_BGR2RGB))
                plt.title(f'Detected Defects ({result["status"]})')
                plt.axis('off')
                
                plt.subplot(1, 3, 3)
                # Create a summary plot
                defect_types = [d['type'] for d in result['defects']]
                if defect_types:
                    unique_types, counts = np.unique(defect_types, return_counts=True)
                    plt.bar(unique_types, counts, color=['red', 'magenta'])
                    plt.title('Defect Type Distribution')
                    plt.ylabel('Count')
                else:
                    plt.text(0.5, 0.5, 'No Defects\nDetected', 
                            ha='center', va='center', fontsize=16, color='green')
                    plt.xlim(0, 1)
                    plt.ylim(0, 1)
                    plt.title('Quality Status')
                    plt.axis('off')
                
                plt.tight_layout()
                plt.show()
                
        except Exception as e:
            print(f"Error processing {image_path}: {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from torchvision.transforms import ToTensor
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
import datetime

# Paths (update these)
model_path = r'C:\Users\krish\OneDrive\Desktop\Hackathon\src\problem_1\defect_detection_model.pth'  # Path to downloaded model
test_images_dir = r"C:\Users\krish\OneDrive\Desktop\Hackathon\src\problem_1\test_dataset"  # Directory with test images
output_dir = r"C:\Users\krish\OneDrive\Desktop\Hackathon\src\problem_1\inference_outputs"  # Directory for visualizations

# Device (CPU is sufficient)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load model
model = fasterrcnn_resnet50_fpn(pretrained=False)
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes=3)  # Background, Cut, Flash
model.load_state_dict(torch.load(model_path, map_location=device))
model.to(device)
model.eval()

# Get list of images
image_files = [f for f in os.listdir(test_images_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
if not image_files:
    raise FileNotFoundError(f"No images found in {test_images_dir}")

# Create output directory
os.makedirs(output_dir, exist_ok=True)

# Process each image
transform = ToTensor()
class_names = ['Background', 'Cut', 'Flash']
confidence_threshold = 0.5

for image_file in image_files:
    image_path = os.path.join(test_images_dir, image_file)
    
    # Load and preprocess image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Warning: Could not load image {image_path}")
        continue
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    image_tensor = transform(image).to(device)
    
    # Inference
    with torch.no_grad():
        prediction = model([image_tensor])[0]
    
    # Process predictions
    boxes = prediction['boxes'].cpu().numpy()
    labels = prediction['labels'].cpu().numpy()
    scores = prediction['scores'].cpu().numpy()
    
    # Classify and localize
    print(f"\nInference on: {image_path} ({datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')})")
    defect_detected = False
    for box, label, score in zip(boxes, labels, scores):
        if score >= confidence_threshold and label in [1, 2]:  # Cut or Flash
            print(f"Result: Defect Class: {class_names[label]}, Score: {score:.2f}, Box: {box}")
            defect_detected = True
    
    if not defect_detected:
        print("Result: Normal (no defect detected)")
    
    # Visualize results
    fig, ax = plt.subplots(1)
    ax.imshow(image)
    if defect_detected:
        for box, label, score in zip(boxes, labels, scores):
            if score >= confidence_threshold and label in [1, 2]:
                rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], 
                                        linewidth=2, edgecolor='r', facecolor='none')
                ax.add_patch(rect)
                ax.text(box[0], box[1], f"{class_names[label]} {score:.2f}", 
                        bbox=dict(facecolor='white', alpha=0.5))
    else:
        ax.text(10, 10, "Normal (no defect)", bbox=dict(facecolor='white', alpha=0.5))
    plt.axis('off')
    
    # Save visualization
    output_path = os.path.join(output_dir, f"result_{image_file}")
    plt.savefig(output_path)
    plt.close()
    print(f"Visualization saved to: {output_path}")
```
